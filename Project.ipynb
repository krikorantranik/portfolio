This is a case study about a data engineering project I loved to conduct not too long ago, which allowed me to choose and implement a complete solution to answer all of an internal customer's data needs. The details have been changed for privacy.

The issue was as follows: the customer was using different data sources, sometimes residing in disconnected networks between the same company and using different technologies to produce reports as needed. Of these, some were on a regular schedule, and some weren't. The leaders were known to reach this department for analytics requiring immediate results, which was complicated, due to the size of the data and the disparate sources of it. Sounds familiar?

The sources were basically:

1. Complaint data (tens of thousands of complaints monthly, across a very rich product portfolio). This data was flawed, full of typos and specific language. The agents could select dropdowns to provide some metadata about the complaint, but they seldom did it accurately.

2. Device information. This was hosted in a different system (SAP), with a primary key of serial number. The complaints should also have a serial number, but it was rarely correct.

3. Machine data: This is the main source of truth, it holds very useful information, such as versions, installed applications, and hardware but it is also stored in a different system, and the serial numbers are about 80-90% accurate.

4. Device log files: These were stored as files. Hadoop was used as Data Warehouse, but the plan was to decommission it.

5. Bugs and production issues are stored as well in a different system. These points and the machine information reside in a different network than the rest of the data sources.

6. For reporting the preferred technology was Tableau, which was on a physical server.

That being said, I proposed using an Azure subscription to collect and store the necessary data from these sources, which will allow the users to retrieve the information swiftly and provide results in the turnaround times requested by leadership.
